```python
from ultralytics import YOLO
from IPython.display import display, Image

# Load your trained model
model = YOLO(r"C:\Users\tejes\OneDrive\Desktop\project\YOLO_Project\runs\detect\train\weights\best.pt")

# Correct image path
test_image = r"C:\Users\tejes\OneDrive\Desktop\project\YOLO_Project\SMOKE.JPG"

# Run prediction
results = model.predict(source=test_image, conf=0.25, save=True)

# Print detected classes
detected_classes = set()
for result in results:
    for box in result.boxes:
        cls_name = result.names[int(box.cls)].lower()
        detected_classes.add(cls_name)

if "fire" in detected_classes and "smoke" in detected_classes:
    print("🔥 Fire and Smoke detected!")
elif "fire" in detected_classes:
    print("🔥 Fire detected!")
elif "smoke" in detected_classes:
    print("💨 Smoke detected!")
else:
    print("❌ Nothing detected!")

# Show the output image
output_img = f"{results[0].save_dir}/{test_image.split('/')[-1]}"
display(Image(filename=output_img))
```

```python
from ultralytics import YOLO
import cv2

# Load your trained model
model = YOLO("runs/detect/train/weights/last.pt")

# Open webcam
cap = cv2.VideoCapture(0)
save_path = "runs/detect/fire_detected"
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection on the current frame
    results = model.predict(frame, conf=0.25, show=False, save=False)
    
    # Check if fire class is detected
    detected_classes = [model.names[int(cls)] for cls in results[0].boxes.cls]
    if "fire" in detected_classes:
        print("🔥 Fire detected! Stopping webcam...")
        # Save the frame where fire was detected
        cv2.imwrite(f"{save_path}/fire_frame_{frame_count}.jpg", frame)
        break

    # Optional: display live webcam feed
    cv2.imshow("Webcam Feed", frame)
    frame_count += 1

    # Press 'q' to exit manually as a backup
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
